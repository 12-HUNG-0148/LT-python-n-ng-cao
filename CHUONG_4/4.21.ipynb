### ý 1

```python
import pandas as pd
drink = pd.read_csv('drinks.csv')
print(type(drink), "là kiểu dữ liệu của drink")
print(drink.shape, "là kích thước của drink")
print(drink.columns, "\nlà các cột của drink")

print(drink.head(), "\nlà 5 dòng đầu")
print(drink.tail(), "\nlà 5 dòng cuối")
```

### ý 2

```python
drink.groupby('continent').agg({
    'beer_servings': 'mean'
})
```

### ý 3

```python
drink.groupby('continent').agg({
    'wine_servings': 'describe'
})
```

### ý 4

```python
drink.groupby('continent').agg({
    'beer_servings': 'mean',
    'wine_servings': 'mean',
    'spirit_servings':'mean'

})
```

### ý 5

```python
drink.groupby('continent').agg({
    'beer_servings': 'median',
    'wine_servings': 'median',
    'spirit_servings':'median'

})
```
